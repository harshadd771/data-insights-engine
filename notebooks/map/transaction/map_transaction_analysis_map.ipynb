# Transaction data analysis #

```python
import pandas as pd
import json
import os

#This is to direct the path to get the data as states
path="D:\\GITHUB REPO\\pulse\\data\\map\\transaction\\hover\\country\\india\\state\\"
map_state_list=os.listdir(path)
map_state_list

#<------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------>#

clm={'state':[], 'year':[],'quarter':[],'region_name':[], 'transaction_count':[], 'transaction_amount':[]}

for i in map_state_list:
    p_i=path+i+"\\"
    map_yr=os.listdir(p_i)
    for j in map_yr:
        p_j=p_i+j+"\\"
        map_yr_list=os.listdir(p_j)
        for k in map_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for m in D['data']['hoverDataList']:
                name = m['name']
                for n in m['metric']:
                    count = n['count']
                    amount = n['amount']
                    clm['region_name'].append(name)
                    clm['transaction_count'].append(count)
                    clm['transaction_amount'].append(amount)
                    clm['state'].append(i)
                    clm['year'].append(j)
                    clm['quarter'].append(int(k.strip('.json')))


map_trans_df=pd.DataFrame(clm)
```

## Create Connection and database ##

```python
import psycopg2

DB_NAME = "smart_pay"
conn_params = {
    "host": "localhost",
    "database": "postgres",
    "user": "postgres",
    "password": "",
}

conn = psycopg2.connect(**conn_params)
conn.autocommit = True
cur = conn.cursor()

cur.execute("SELECT 1 FROM pg_database WHERE datname = %s", (DB_NAME,))
exists = cur.fetchone()

if not exists:
    cur.execute(f'CREATE DATABASE "{DB_NAME}"')
    print(f"Database '{DB_NAME}' created successfully")
else:
    print(f"Database '{DB_NAME}' already exists")

cur.close()
conn.close()
```

## Create Map_Transaction Table In Database ##

```python
import psycopg2

conn = psycopg2.connect(
    host="localhost",
    database="smart_pay",
    user="postgres",
    password="",
)
conn.autocommit = True
cur = conn.cursor()

cur.execute("""
    CREATE TABLE IF NOT EXISTS map_transaction (
        id SERIAL PRIMARY KEY,
        state VARCHAR(255) NOT NULL,
        year INTEGER NOT NULL,
        quarter INTEGER,
        region_name VARCHAR(255),
        transaction_count BIGINT,
        transaction_amount NUMERIC(15, 2)
    );
""")

cur.execute("SELECT to_regclass('public.map_transaction')")
print("Table exists in PostgreSQL:", cur.fetchone()[0])

conn.commit()
cur.close()
conn.close()
print("Table 'map_transaction' created successfully")
```

## Inserting Data from Data Frame to Transaction_data Table ##

```python
import psycopg2

conn = psycopg2.connect(
    host="localhost",
    database="smart_pay",
    user="postgres",
    password="",
)
conn.autocommit = False
cur = conn.cursor()

for _, row in map_trans_df.iterrows():
    cur.execute(
        """
        INSERT INTO map_transaction (
            state,
            year,
            quarter,
            region_name,
            transaction_count,
            transaction_amount
        )
        VALUES (%s, %s, %s, %s, %s, %s)
        """,
        (
            row["state"],
            row["year"],
            row["quarter"],
            row["region_name"],
            row["transaction_count"],
            row["transaction_amount"],
        ),
    )

conn.commit()
cur.close()
conn.close()
print("Data inserted successfully")
```
