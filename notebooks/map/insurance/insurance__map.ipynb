# Insurance data analysis #

```python
import pandas as pd
import json
import os

#This is to direct the path to get the data as states
path="D:\\GITHUB REPO\\pulse\\data\\map\\insurance\\hover\\country\\india\\state\\"
map_state_list=os.listdir(path)
map_state_list

clm={'state':[], 'year':[],'quarter':[],'region_name':[], 'transaction_count':[], 'transaction_amount':[]}

for i in map_state_list:
    p_i=path+i+"\\"
    map_yr=os.listdir(p_i)
    for j in map_yr:
        p_j=p_i+j+"\\"
        map_yr_list=os.listdir(p_j)
        for k in map_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for m in D['data']['hoverDataList']:
                name = m['name']
                for n in m['metric']:
                    count = n['count']
                    amount = n['amount']
                    clm['region_name'].append(name)
                    clm['transaction_count'].append(count)
                    clm['transaction_amount'].append(amount)
                    clm['state'].append(i)
                    clm['year'].append(j)
                    clm['quarter'].append(int(k.strip('.json')))


map_insurance_df=pd.DataFrame(clm)
```

## Create Map_Insurance Table In Database ##

```python
import psycopg2

conn = psycopg2.connect(
    host="localhost",
    database="smart_pay",
    user="postgres",
    password="HareKrishna@143",
)
conn.autocommit = True
cur = conn.cursor()

cur.execute("""
    CREATE TABLE IF NOT EXISTS map_insurance (
        id SERIAL PRIMARY KEY,
        state VARCHAR(255) NOT NULL,
        year INTEGER NOT NULL,
        quarter INTEGER,
        region_name VARCHAR(255),
        transaction_count BIGINT,
        transaction_amount NUMERIC(15, 2)
    );
""")

cur.execute("SELECT to_regclass('public.map_insurance')")
print("Table exists in PostgreSQL:", cur.fetchone()[0])

conn.commit()
cur.close()
conn.close()
print("Table 'map_insurance' created successfully")
```

## Inserting Data from Data Frame to Transaction_data Table ##

```python
import psycopg2

conn = psycopg2.connect(
    host="localhost",
    database="smart_pay",
    user="postgres",
    password="HareKrishna@143",
)
conn.autocommit = False
cur = conn.cursor()

for _, row in map_insurance_df.iterrows():
    cur.execute(
        """
        INSERT INTO insurance_map (
            state,
            year,
            quarter,
            region_name,
            transaction_count,
            transaction_amount
        )
        VALUES (%s, %s, %s, %s, %s, %s)
        """,
        (
            row["state"],
            row["year"],
            row["quarter"],
            row["region_name"],
            row["transaction_count"],
            row["transaction_amount"],
        ),
    )

conn.commit()
cur.close()
conn.close()
print("Data inserted successfully")
```
