# User data analysis #

```python
import pandas as pd
import json
import os

#This is to direct the path to get the data as states
path="D:\\GITHUB REPO\\pulse\\data\\map\\user\\hover\\country\\india\\state\\"
map_state_list=os.listdir(path)
map_state_list

clm={'state':[], 'year':[],'quarter':[],'region_name':[], 'registeredUsers':[], 'appOpens':[]}

for i in map_state_list:
    p_i=path+i+"\\"
    map_yr=os.listdir(p_i)
    for j in map_yr:
        p_j=p_i+j+"\\"
        map_yr_list=os.listdir(p_j)
        for k in map_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            hover_data = D['data']['hoverData']
            for region_name, values in hover_data.items():
                clm['region_name'].append(region_name)
                clm['registeredUsers'].append(values['registeredUsers'])
                clm['appOpens'].append(values['appOpens'])
                clm['state'].append(i)
                clm['year'].append(j)
                clm['quarter'].append(int(k.strip('.json')))


map_user_df=pd.DataFrame(clm)
```

## Create Map_User Table In Database ##

```python
import psycopg2

conn = psycopg2.connect(
    host="localhost",
    database="smart_pay",
    user="postgres",
    password="HareKrishna@143",
)
conn.autocommit = True
cur = conn.cursor()

cur.execute("""
    CREATE TABLE IF NOT EXISTS user_map (
        id SERIAL PRIMARY KEY,
        state VARCHAR(255) NOT NULL,
        year INTEGER NOT NULL,
        quarter INTEGER,
        region_name VARCHAR(255),
        registeredUsers BIGINT,
        appOpens NUMERIC(15, 2)
    );
""")

cur.execute("SELECT to_regclass('public.user_map')")
print("Table exists in PostgreSQL:", cur.fetchone()[0])

conn.commit()
cur.close()
conn.close()
print("Table 'user_map' created successfully")
```

## Inserting Data from Data Frame to user_map Table ##

```python
import psycopg2

conn = psycopg2.connect(
    host="localhost",
    database="smart_pay",
    user="postgres",
    password="HareKrishna@143",
)
conn.autocommit = False
cur = conn.cursor()

for _, row in map_user_df.iterrows():
    cur.execute(
        """
        INSERT INTO user_map (
            state,
            year,
            quarter,
            region_name,
            registeredUsers,
            appOpens
        )
        VALUES (%s, %s, %s, %s, %s, %s)
        """,
        (
            row["state"],
            row["year"],
            row["quarter"],
            row["region_name"],
            row["registeredUsers"],
            row["appOpens"],
        ),
    )

conn.commit()
cur.close()
conn.close()
print("Data inserted successfully")
```
