# Top Users Pincodes data analysis #

```python
import pandas as pd
import json
import os

#This is to direct the path to get the data as states
path="D:\\GITHUB REPO\\pulse\\data\\top\\user\\country\\india\\state\\"
top_state_list=os.listdir(path)
top_state_list
clm={'state':[], 'year':[],'quarter':[],'pincodes':[], 'registeredUsers':[] }

for i in top_state_list:
    p_i=path+i+"\\"
    top_yr=os.listdir(p_i)
    for j in top_yr:
        p_j=p_i+j+"\\"
        top_yr_list=os.listdir(p_j)
        for k in top_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']['pincodes']:
                pincodes = z['name']
                registeredUsers=z['registeredUsers']
                clm['pincodes'].append(pincodes)
                clm['registeredUsers'].append(registeredUsers)
                clm['state'].append(i)
                clm['year'].append(j)
                clm['quarter'].append(int(k.strip('.json')))

top_users_pincodes_df=pd.DataFrame(clm)

top_users_pincodes_df
```

## Create users_pincodes_top Table In Database ##

```python
import psycopg2

conn = psycopg2.connect(
    host="localhost",
    database="smart_pay",
    user="postgres",
    password="",
)
conn.autocommit = True
cur = conn.cursor()

cur.execute("""
    CREATE TABLE IF NOT EXISTS users_pincodes_top (
        id SERIAL PRIMARY KEY,
        state VARCHAR(255) NOT NULL,
        year INTEGER NOT NULL,
        quarter INTEGER NOT NULL,
        pincodes VARCHAR(255) NOT NULL,
        registeredUsers BIGINT
    );
""")

cur.execute("SELECT to_regclass('public.users_pincodes_top')")
print("Table exists in PostgreSQL:", cur.fetchone()[0])

conn.commit()
cur.close()
conn.close()
print("Table 'users_pincodes_top' created successfully")
```

## Inserting Data from Data Frame to Users_pincodes_Top Table ##

```python
import psycopg2

conn = psycopg2.connect(
    host="localhost",
    database="smart_pay",
    user="postgres",
    password="",
)
conn.autocommit = False
cur = conn.cursor()

for _, row in top_users_pincodes_df.iterrows():
    cur.execute(
        """
        INSERT INTO users_pincodes_top (
            state,
            year,
            quarter,
            pincodes,
            registeredUsers
        )
        VALUES (%s, %s, %s, %s, %s)
        """,
        (
            row["state"],
            row["year"],
            row["quarter"],
            row["pincodes"],
            row["registeredUsers"],
        ),
    )

conn.commit()
cur.close()
conn.close()
print("Data inserted successfully")
```
