# Top Transaction Districts data analysis #

```python
import pandas as pd
import json
import os

#This is to direct the path to get the data as states
path="D:\\GITHUB REPO\\pulse\\data\\top\\transaction\\country\\india\\state\\"
top_state_list=os.listdir(path)
top_state_list
clm={'state':[], 'year':[],'quarter':[],'districts':[], 'total_amount':[], 'total_count':[] }

for i in top_state_list:
    p_i=path+i+"\\"
    top_yr=os.listdir(p_i)
    for j in top_yr:
        p_j=p_i+j+"\\"
        top_yr_list=os.listdir(p_j)
        for k in top_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']['districts']:
                districts = z['entityName']
                total_amount = z['metric']['amount']
                total_count = z['metric']['count']
                clm['districts'].append(districts)
                clm['total_amount'].append(total_amount)
                clm['total_count'].append(total_count)
                clm['state'].append(i)
                clm['year'].append(j)
                clm['quarter'].append(int(k.strip('.json')))

top_transactions_districts_df=pd.DataFrame(clm)

top_transactions_districts_df
```

## Create users_district_top Table In Database ##

```python
import psycopg2

conn = psycopg2.connect(
    host="localhost",
    database="smart_pay",
    user="postgres",
    password="HareKrishna@143",
)
conn.autocommit = True
cur = conn.cursor()

cur.execute("""
    CREATE TABLE IF NOT EXISTS transaction_districts_top (
        id SERIAL PRIMARY KEY,
        state VARCHAR(255) NOT NULL,
        year INTEGER NOT NULL,
        quarter INTEGER NOT NULL,
        districts VARCHAR(255) NOT NULL,
        total_count BIGINT,
        total_amount NUMERIC(15, 2)
    );
""")

cur.execute("SELECT to_regclass('public.transaction_districts_top')")
print("Table exists in PostgreSQL:", cur.fetchone()[0])

conn.commit()
cur.close()
conn.close()
print("Table 'transaction_districts_top' created successfully")
```

## Inserting Data from Data Frame to Transaction_district_Top Table ##

```python
import psycopg2

conn = psycopg2.connect(
    host="localhost",
    database="smart_pay",
    user="postgres",
    password="HareKrishna@143",
)
conn.autocommit = False
cur = conn.cursor()

for _, row in top_transactions_districts_df.iterrows():
    cur.execute(
        """
        INSERT INTO transaction_districts_top (
            state,
            year,
            quarter,
            districts,
            total_count,
            total_amount
        )
        VALUES (%s, %s, %s, %s, %s, %s)
        """,
        (
            row["state"],
            row["year"],
            row["quarter"],
            row["districts"],
            row["total_count"],
            row["total_amount"],
        )
    )

conn.commit()
cur.close()
conn.close()
print("Data inserted successfully")

cur.close()
conn.close()
print("Data inserted successfully")
```
