# User Data Analysis
Notebook for exploring user-related datasets.

```python
import pandas as pd
import json
import os

#This is to direct the path to get the data as states
path="D:\\GITHUB REPO\\pulse\\data\\aggregated\\user\\country\\india\\state\\"
Agg_user_list=os.listdir(path)
Agg_user_list

#<------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------>#

#This is to extract the data's to create a dataframe

clm = { 'State': [], 'Year': [], 'Quarter': [],'Registered_Users': [], 'App_Opens': [],'brand': [], 'count': [], 'percentage': []}

for i in Agg_user_list:
    p_i = path + i + "\\"
    Agg_yr = os.listdir(p_i)
    for j in Agg_yr:
        p_j = p_i + j + "\\"
        Agg_yr_list = os.listdir(p_j)
        for k in Agg_yr_list:
            p_k = p_j + k
            with open(p_k, 'r', encoding="utf-8") as Data:
                D = json.load(Data)

            Registered_Users = D['data']['aggregated']['registeredUsers']
            App_Opens = D['data']['aggregated']['appOpens']

            if D['data']['usersByDevice'] is not None:
                for y in D['data']['usersByDevice']:
                    clm['State'].append(i)
                    clm['Year'].append(j)
                    clm['Quarter'].append(int(k.strip('.json')))
                    clm['Registered_Users'].append(Registered_Users)
                    clm['App_Opens'].append(App_Opens)
                    clm['brand'].append(y['brand'])
                    clm['count'].append(y['count'])
                    clm['percentage'].append(y['percentage'])
            else:
                clm['State'].append(i)
                clm['Year'].append(j)
                clm['Quarter'].append(int(k.strip('.json')))
                clm['Registered_Users'].append(Registered_Users)
                clm['App_Opens'].append(App_Opens)
                clm['brand'].append(None)
                clm['count'].append(None)
                clm['percentage'].append(None)

Agg_user_df = pd.DataFrame(clm)

```

## TO CHECK OUTLIERS ON app opens with registered users 
“In West Bengal, during Q4 of 2024, App Opens = 1,994,940,763. This value is much higher compared to the other quarters, so it will be flagged as an outlier.” ###

```python
Q1 = Agg_user_df['app_opens'].quantile(0.25)
Q3 = Agg_user_df['app_opens'].quantile(0.75)
IQR = Q3 - Q1

outliers = Agg_user_df[
    (Agg_user_df['app_opens'] < (Q1 - 1.5 * IQR)) |
    (Agg_user_df['app_opens'] > (Q3 + 1.5 * IQR))
]
print(outliers)
```

```python

Q1 = Agg_user_df['app_opens'].quantile(0.25)
Q3 = Agg_user_df['app_opens'].quantile(0.75)
IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR


Agg_user_df['is_data_issue'] = (
    ((Agg_user_df['registered_users'] > 0) & (Agg_user_df['app_opens'] == 0))   # Rule 1
    |
    (Agg_user_df['app_opens'] < Agg_user_df['registered_users'])                # Rule 2
    |
    ((Agg_user_df['app_opens'] < lower_limit) | (Agg_user_df['app_opens'] > upper_limit))  # Rule 3
)

print(Agg_user_df[['state','year','quarter','app_opens','is_data_issue']])
```

## CREATE CONNECTION DATABASE AND CREATE TABLE ##

```python
import psycopg2

conn = psycopg2.connect(
    host="localhost",
    database="smart_pay",
    user="postgres",
    password="",
)
conn.autocommit = True
cur = conn.cursor()

cur.execute("""
    CREATE TABLE IF NOT EXISTS user_data (
        id SERIAL PRIMARY KEY,
        state VARCHAR(255) NOT NULL,
        year INTEGER NOT NULL,
        quarter INTEGER,
        registered_users BIGINT,
        app_opens BIGINT,
        brand VARCHAR(255) NOT NULL,
        count BIGINT,
        percentage DECIMAL(5,4)
    );
""")

conn.commit()
cur.close()
conn.close()
print("Table 'user_data' created successfully")
```

## ADDING is_data_issue column in table ##

```python
import psycopg2

conn = psycopg2.connect(
    host="localhost",
    database="smart_pay",
    user="postgres",
    password="",
)
conn.autocommit = True
cur = conn.cursor()

cur.execute("""
    ALTER TABLE user_data 
    ADD COLUMN is_data_issue BOOLEAN;
""")

conn.commit()
cur.close()
conn.close()
print("COLUMN Added successfully")
```

## INSERT DATA INTO USER_DATA TABLE ##

```python
import psycopg2

conn = psycopg2.connect(
    host="localhost",
    database="smart_pay",
    user="postgres",
    password="",
)
conn.autocommit = False
cur = conn.cursor()

for _, row in Agg_user_df.iterrows():
    cur.execute(
        """
        INSERT INTO user_data (
            state,
            year,
            quarter,
            registered_users,
            app_opens,
            brand,
            count,
            percentage,
            is_data_issue
        )
        VALUES (%s, %s, %s, %s, %s, %s,%s, %s, %s)
        """,
        (
            row["state"],
            row["year"],
            row["quarter"],
            row["registered_users"],
            row["app_opens"],
            row["brand"],
            row["count"],
            row["percentage"],
            row["is_data_issue"]
        ),
    )

conn.commit()
cur.close()
conn.close()
print("Data inserted successfully")
```
