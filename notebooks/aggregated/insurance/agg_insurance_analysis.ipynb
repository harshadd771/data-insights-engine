# Aggregated Insurance data analysis #

```python
import pandas as pd
import json
import os

#This is to direct the path to get the data as states
path="D:\\GITHUB REPO\\pulse\\data\\aggregated\\insurance\\country\\india\\state\\"
Agg_state_list=os.listdir(path)
Agg_state_list
clm={'state':[], 'year':[],'quarter':[],'data_from':[], 'data_to':[], 'payment_category':[], 'total_count':[], 'total_amount':[]}

for i in Agg_state_list:
    p_i=path+i+"\\"
    Agg_yr=os.listdir(p_i)
    for j in Agg_yr:
        p_j=p_i+j+"\\"
        Agg_yr_list=os.listdir(p_j)
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            data_from = D['data']['from']
            data_to = D['data']['to']
            for z in D['data']['transactionData']:
              payment_category = z['name']
              for m in z['paymentInstruments']:
                count=m['count']
                amount=m['amount']
                clm['data_from'].append(data_from)
                clm['data_to'].append(data_to)
                clm['payment_category'].append(payment_category)
                clm['total_count'].append(count)
                clm['total_amount'].append(amount)
                clm['state'].append(i)
                clm['year'].append(j)
                clm['quarter'].append(int(k.strip('.json')))

agg_insurance_df=pd.DataFrame(clm)

agg_insurance_df
```

## Create Connection and database ##

```python
import psycopg2

DB_NAME = "smart_pay"
conn_params = {
    "host": "localhost",
    "database": "postgres",
    "user": "postgres",
    "password": "",
}

conn = psycopg2.connect(**conn_params)
conn.autocommit = True
cur = conn.cursor()

cur.execute("SELECT 1 FROM pg_database WHERE datname = %s", (DB_NAME,))
exists = cur.fetchone()

if not exists:
    cur.execute(f'CREATE DATABASE "{DB_NAME}"')
    print(f"Database '{DB_NAME}' created successfully")
else:
    print(f"Database '{DB_NAME}' already exists")

cur.close()
conn.close()
```

```python
import psycopg2

conn = psycopg2.connect(
    host="localhost",
    database="smart_pay",
    user="postgres",
    password="",
)
conn.autocommit = True
cur = conn.cursor()

cur.execute("""
    CREATE TABLE IF NOT EXISTS insurance_agg (
        id SERIAL PRIMARY KEY,
        state VARCHAR(255) NOT NULL,
        year INTEGER NOT NULL,
        quarter INTEGER NOT NULL,
        data_from TIMESTAMP NOT NULL,
        data_to TIMESTAMP NOT NULL,
        payment_category VARCHAR(255),
        total_count BIGINT,
        total_amount NUMERIC(15, 2)
    );
""")

cur.execute("SELECT to_regclass('public.insurance_agg')")
print("Table exists in PostgreSQL:", cur.fetchone()[0])

conn.commit()
cur.close()
conn.close()
print("Table 'insurance_agg' created successfully")
```

```python
import psycopg2

conn = psycopg2.connect(
    host="localhost",
    database="smart_pay",
    user="postgres",
    password="",
)
conn.autocommit = False
cur = conn.cursor()

for _, row in agg_insurance_df.iterrows():
    cur.execute(
        """
        INSERT INTO insurance_agg (
            state,
            year,
            quarter,
            data_from,
            data_to,
            payment_category,
            total_count,
            total_amount
        )
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
        """,
        (
            row["state"],
            row["year"],
            row["quarter"],
            row["data_from"],
            row["data_to"],
            row["payment_category"],
            row["total_count"],
            row["total_amount"],
        ),
    )

conn.commit()
cur.close()
conn.close()
print("Data inserted successfully")
```
